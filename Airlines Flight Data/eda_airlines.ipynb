```python
# Menampilkan semua maskapai penerbangan beserta frekuensinya (berapa kali muncul di datasheet)
datasheet['airline'].value_counts()
```

```python
import matplotlib.pyplot as plt

datasheet['airline'].value_counts(ascending=True).plot.bar(
    color=['lightgreen', 'lightblue', 'purple' ]
)
plt.title('Frekuensi Maskapai')
plt.xlabel('Airlines')
plt.ylabel('Number of Flights')
plt.xticks(rotation=0)
plt.show()
```

```python
# Visualisasi menggunakan seaborn
import seaborn as sns
airline_counts = datasheet['airline'].value_counts().reset_index()
airline_counts.columns = ['airline','count']

plt.figure(figsize=(10,6))
sns.barplot(x='airline',y='count', data=airline_counts, hue='airline', palette=['lightgreen', 'lightblue','purple'],legend=False)

plt.title('Airlines with Frequencies')
plt.xlabel('Airlines')
plt.ylabel('Number of Flights')
plt.xticks(rotation=0)
plt.show()
```

```python
# Menampilkan Departure Time untuk flight
datasheet['departure_time'].value_counts()
```

```python
# Menampilkan Arrival Time untuk flight
datasheet['arrival_time'].value_counts()
```

```python
plt.figure(figsize= (16,4))
plt.subplot(1,2,1)

# Visualisasi Departure Time
plt.bar( datasheet['departure_time'].value_counts().index, datasheet['departure_time'].value_counts().values, color = ['r','b'])
plt.title('Departure Time')
plt.xlabel('D. Time')
plt.ylabel('D. Freq')

# Visualisasi Arrival Time
plt.subplot(1,2,2)
plt.bar(datasheet['arrival_time'].value_counts().index, datasheet['arrival_time'].value_counts().values,color = ['g','y'])
plt.title('Arrival Time')
plt.xlabel('A. Time')
plt.ylabel('A. Freq')

plt.show()
```

```python
# Menampilkan asal dari setiap penerbangan 
datasheet['source_city'].value_counts()
```

```python
# Menampilkan tujuan dari setiap penerbangan
datasheet['destination_city'].value_counts()
```

```python
# Visualisasi Destination dan Source City

plt.figure(figsize=(16,4))
plt.subplot(1,2,1)

plt.barh( datasheet['destination_city'].value_counts().index, datasheet['destination_city'].value_counts().values, color =['y','r'])
plt.title('Destination Cities with No. of flight')
plt.ylabel('Cities')
plt.xlabel('No of Flights')

plt.subplot(1,2,2)
plt.barh( datasheet['source_city'].value_counts().index, datasheet['source_city'].value_counts().values, color = ['g','m'])
plt.title('Source Cities with No. of flight')
plt.ylabel('Cities')
plt.xlabel('No of Flights')

plt.show()
```

```python
# Menghitung rata-rata harga tiket untuk setiap maskapai
datasheet.groupby('airline')['price'].mean()
```

```python
# Membuat tampilan visualisasi rata-rata harga tiket untuk beberapa airlines
sns.catplot(x= 'airline', y='price', kind='bar', palette='rocket', data = datasheet, hue='class', height=5, weights=2)
```

```python
# Menampilkan harga tiket termahal berdasarkan waktu penerbangan
datasheet.groupby('departure_time')['price'].mean()
```

```python
# Membuat tampilan visualisasi harga tiket berdasarkan waktu penerbangan
sns.catplot(x='departure_time', 
            y='price', 
            kind='bar', 
            palette='rocket', 
            data=datasheet, 
            hue='class', 
            height = 5,
            aspect= 3)
```

```python
# Menampilkan tampilan visualisasi harga tiket berdasarkan waktu kedatangan
sns.catplot(x='arrival_time',
            y= 'price',
            kind='bar',
            palette='rocket',
            data = datasheet,
            hue = 'class',
            height=5,
            aspect=3)
```

```python
# Membuat line plot harga tiket berdasarkan waktu kedatangan, dengan subplot terpisah untuk setiap periode kedatangan
sns.relplot( x = 'arrival_time',
           y = 'price',
           data = datasheet,
           col = 'departure_time',
           kind= 'line')
plt.show()
```
