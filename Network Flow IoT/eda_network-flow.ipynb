```python
import matplotlib.pyplot as plt
import seaborn as sns

df['src'].value_counts(ascending=True).plot.barh(
    color=['lightgreen', 'purple']
)
plt.title('Frekuensi src request')
plt.xlabel('Frekuensi muncul')
plt.ylabel('IP Address Sender')
plt.show()
```

```python
attack_src = df[df['label']==1]['src'].value_counts()

plt.figure(figsize=(14,8), dpi=100)
attack_src.plot(kind='barh', color='blue')

# Tambah label angka di samping bar
for idx, val in enumerate(attack_src.values):
    plt.text(val, idx, str(val), color='red', va='center', fontsize=10)

plt.xlabel('Number of Requests')
plt.ylabel('IP Address of Sender')
plt.title('Number of Attack Requests per Source IP')
plt.subplots_adjust(left=0.3)  # kasih ruang label IP
plt.show()
```

```python
attack_src = df[df['label']==0]['src'].value_counts()

plt.figure(figsize=(14,8), dpi=100)
attack_src.plot(kind='barh', color='yellow')

# Tambah label angka di samping bar
for idx, val in enumerate(attack_src.values):
    plt.text(val, idx, str(val), color='red', va='center', fontsize=10)

plt.xlabel('Number of Requests')
plt.ylabel('IP Address of Sender')
plt.title('Number of Normal Requests per Source IP')
plt.subplots_adjust(left=0.3)  # kasih ruang label IP
plt.show()
```
